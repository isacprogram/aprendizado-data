# CONSTRUÇÃO DE BOXSPLOT E HISTOGRAMAS EM BASE DE NACIDOS VIVOS DO SUS NO ANO DE 2019

```python
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
%matplotlib inline
#matplot notebook
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Abrindo base Sinasc
sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc
```

```python
# selecionando as variáveis de interesse e tratamento otimizado de dados faltantes
df_i = sinasc[['IDADEMAE','PESO']]
df_i.dropna()
df_i
```

```python
# verificandos dados faltantes
df_i.isna().sum()
```

```python
# realizando boxplot na variável idade da mãe através do pandas
df_i[['IDADEMAE']].boxplot()
```

####  2) Faça um boxplot do peso do bebê

```python
#realizando boxsplot da variável peso pelo seaborn
sns.boxplot(y='PESO', data=df_i)
```

####  3) Faça um histograma da idade da mãe

```python
#verificando histograma na variável discreta idade da mãe.

df_i[['IDADEMAE']].hist(bins=20)
```

#### 4) Faça um histograma do peso do bebê

```python
# verificando histograma de váriavel continua peso do RN. alpha=tranparência,bins=tamanho,kde=linha.


sns.displot(x='PESO',bins=20,alpha=.25,kde=True, data=df_i)
```
